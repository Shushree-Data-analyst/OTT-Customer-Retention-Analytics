# Check Missing Values

```python
ott.isnull().sum()
```

```python
sns.boxplot(x=ott['monthly_charge_inr'])
plt.title('Boxplot of Monthly Charge INR')
plt.show()
```

# Handle Missing Values

```python
ott['city'] = ott['city'].fillna(ott['city'].mode()[0])
```

```python
ott['monthly_charge_inr'] = ott['monthly_charge_inr'].fillna(ott['monthly_charge_inr'].median())
```

```python
ott['nps_score'] = ott['nps_score'].fillna(ott['nps_score'].median())
```

```python
ott.isnull().sum()
```

```python
for i in ott.columns:
    print(i,":","\n",ott[i].unique(),"\n") 
```

```python
ott['avg_watch_hrs_week'] = ott['avg_watch_hrs_week'].abs()
```

# Exploratory Data Analysis (EDA)

```python
sns.countplot(x='churned', data=ott)
plt.title('Customer Churn Distribution')
plt.show()
```

# Churn by Device Type

```python
sns.countplot(x='primary_device', hue='churned', data=ott)
plt.xticks(rotation=45)
plt.title('Churn by Device Type')
plt.show()
```

# Churn by Payment Method

```python
sns.countplot(x='payment_method', hue='churned', data=ott)
plt.xticks(rotation=45)
plt.title('Churn by Payment Method')
plt.show()
```

# NPS Score Analysis

```python
 sns.boxplot(x='churned', y='nps_score', data=ott)
plt.title('NPS Score vs Churn')

plt.show()
```

# Churn by Different Age Group

```python
sns.countplot(x='age_group', hue='churned', data=ott)

plt.title('Churn by Age Group')

plt.xlabel('Age Group')

plt.ylabel('Customer Count')

plt.show()
```
